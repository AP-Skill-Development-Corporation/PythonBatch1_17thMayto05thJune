Matplotlib 
---------------
- Matplotlib is one of the library in python for constructing 2D plotting Visualiztionn.
- it is used to create high quality graphs, plots,charts, figures,etc

### Pyplot
- Pyplot is one of the popular module in matplotlib library and it contains collection of functions, which is a simple and easy user interface for constructing plots

```python
import matplotlib.pyplot as plt
```

```python
x = [10,12,13,15]
y = [5,6,7,8]
plt.plot(x,y)
plt.show()
```

```python
x = [10,12,14,16]
y = [5,6,7,8]
plt.plot(x,y,label = "Line",color="red")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line plot")
plt.legend(loc="upper left")
```

```python
import numpy as np
x = np.arange(20)
x
```

```python
plt.plot(x,x**2,color = "red",label = "squares")
plt.plot(x,x**3,color = "green",label = "cubes")
plt.plot(x,x**4,color = "yellow",label = "four")
plt.legend()
```

```python
plt.plot(x,x+10,color = 'g',linestyle = '--')
plt.plot(x,x+20,color = 'm',linestyle = '-.')
```

```python
year = ['2019 Dec','2020 Jan','2020 March','2020 july']
goldprice = [10000,20000,30000,40000]
plt.bar(year,goldprice,color='m')
```

```python
x = np.random.randint(1000,1500,30)
y = np.random.randint(500,1000,30)
plt.scatter(x,y)
```

```python
games = ["football","kabbadi","cricket","chess"]
score = [80,92,75,50]
plt.pie(score,labels = games)
```

```python
games = ["football","kabbadi","cricket","chess"]
score = [80,92,75,50]
plt.pie(score,labels = games,explode =(0,0.2,0,0),startangle = 90,colors = ["c","g","m","pink"],shadow = True)
plt.show()
plt.savefig('piechart.png')
```

```python
h = [0,0.5,0.6,1.5,1.6,3.6,3.7,3.8,5.5,5.1,5.2]
bins = 4
plt.hist(h,bins,rwidth = 0.4)
```

```python
days = [1,2,3,4,5]
age = [40,72,92,11,35]
weight = [15,30,70,48,60]
plt.stackplot(days,age,weight,colors = ['c','g'])
plt.show()
```
